```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020
```

```python
# Display the 2021 sales DataFrame
sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print(sales_2020.dtypes)

print(sales_2020.info())
```

```python
# Check the 2021 sales data types.
print(sales_2021.dtypes)

print(sales_2021.info())
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([sales_2020, sales_2021], ignore_index = True)

combined_sales.reset_index(drop = True, inplace = True) 

combined_sales
```

```python
# Check if any values are null.
print(combined_sales.isnull().sum())
```

```python
# Check the data type of each column
print(combined_sales.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype

print(combined_sales['invoice_date'].unique())

combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'], errors='coerce')
print(combined_sales[combined_sales['invoice_date'].isnull()]) #check for any roles if conversion failed or did not happen

combined_sales

#combined_sales['invoice_date']= pd.to_datetime(combined_sales['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_sales.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

region_sales_groupby = combined_sales.groupby(['region', 'state', 'city']).agg({'units_sold':'sum'})
region_sales_groupby.rename(columns={'units_sold':'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
top_regions_groupby = region_sales_groupby.sort_values(by = 'Total_Products_Sold',ascending=False).head(5)
top_regions_groupby
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
region_sales_pivot = combined_sales.pivot_table(
    values = 'units_sold', 
    index=['region','state','city'], 
    aggfunc = 'sum')

# Rename the "units_sold" column to "Total_Products_Sold"

region_sales_pivot.rename(columns={'units_sold':'Total_Products_Sold'},inplace=True)


# Show the top 5 results.
top_regions_pivot = region_sales_pivot.sort_values(by = 'Total_Products_Sold', ascending=False).head(5)
top_regions_pivot
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_groupby = combined_sales.groupby(['region','state','city']).agg({'total_sales':'sum'})

total_sales_groupby.rename(columns={'total_sales':'Total Sales'}, inplace=True)

# Show the top 5 results.
top_total_sales_groupby = total_sales_groupby.sort_values(by = 'Total Sales', ascending=False).head(5)
top_total_sales_groupby
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

total_sales_pivot = combined_sales.pivot_table(
    values = 'total_sales',
    index=['region','state','city'],
    aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot.rename(columns={'total_sales':'Total Sales'}, inplace=True)


# Show the top 5 results.
top_total_sales_pivot = total_sales_pivot.sort_values(by = 'Total Sales', ascending = False).head(5)
top_total_sales_pivot
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"

retailer_sales_groupby = combined_sales.groupby(['retailer','region','state','city']).agg({'total_sales':'sum'})

retailer_sales_groupby.rename(columns={'total_sales':'Total Sales'}, inplace=True)


# Show the top 5 results.
top_retailer_sales_groupby = retailer_sales_groupby.sort_values(by = 'Total Sales', ascending=False).head(5)
top_retailer_sales_groupby
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_pivot = combined_sales.pivot_table(
    values='total_sales', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
retailer_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the top 5 results.
top_retailer_sales_pivot = retailer_sales_pivot.sort_values(by='Total Sales', ascending=False).head(5)
top_retailer_sales_pivot
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = combined_sales[combined_sales['product'].str.contains("Women's Athletic Footwear")]

womens_footwear_sales
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

womens_footwear_groupby = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city']).agg({'units_sold': 'sum'})

womens_footwear_groupby.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

top_womens_footwear_groupby = womens_footwear_groupby.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)

# Show the top 5 results.
top_womens_footwear_groupby
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

womens_footwear_pivot = womens_footwear_sales.pivot_table(
    values='units_sold', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_pivot.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'},inplace=True)


# Show the top 5 results.
top_womens_footwear_pivot = womens_footwear_pivot.sort_values(by='Womens_Footwear_Units_Sold',ascending=False).head(5)

top_womens_footwear_pivot
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

daily_sales_pivot = womens_footwear_sales.pivot_table(
    values='total_sales',
    index='invoice_date',
    aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
daily_sales_pivot.rename(columns={'total_sales':'Total Sales'}, inplace=True)

# Show the table.
daily_sales_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

daily_sales_resampled = daily_sales_pivot.resample('D').sum()


# Sort the resampled pivot table in descending order on "Total Sales".

top_daily_sales = daily_sales_resampled.sort_values(by ='Total Sales', ascending=False).head(10)

top_daily_sales
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales_resampled = daily_sales_pivot.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
top_weekly_sales = weekly_sales_resampled.sort_values(by='Total Sales', ascending=False).head(10)

top_weekly_sales
```
